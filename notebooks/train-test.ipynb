```python
from hilbert import quantum
from hilbert.curves import lib
```

```python
system = quantum.ToyTrain(-50, 50, 101)
gaussian = system.space(lib.Gaussian(1, -1/16, 20, pole=-35))
histories = system.detection_histories(gaussian, x=0)
```

```python
histories.simulate()
```

```python
histories.simulate_density()
```

```python
histories.show_distribution()
```

```python
assert round(histories.weight, 13) == 1, histories.weight
```

```python
assert histories.consistent
```
