```python
from hilbert import quantum
from hilbert.curves import lib
```

```python
system = quantum.ToyTrain(-200, 200, 401)
std = 10
gaussian = system[0](lib.Gaussian(1, -1/(4*std**2), 20, pole=-120))
histories = system.detection_histories(gaussian, x=0)
```

```python
histories.simulate_density(step=2)
```

```python
histories.show_distribution(style='-o')
```

```python
assert round(histories.weight, 13) == 1, histories.weight
```

```python
assert histories.consistent
```

```python
assert system.hop_op.is_unitary()
```
